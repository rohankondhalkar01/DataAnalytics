```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Read the Facebook dataset from a CSV file into a pandas DataFrame
df = pd.read_csv("/Users/rohankondhalkar/Desktop/Pandas/fb.csv")
# Display the entire DataFrame
df
```

```python
# Shift all values in the dataframe down by 1 position
# This creates a new dataframe where each row contains values from the previous row
# The first row will contain NaN values since there's no previous row
# Useful for calculating differences between consecutive rows or creating lag features
df.shift(1)
```

```python
# Shift all values in the dataframe down by 2 positions
# This creates a new dataframe where each value appears 2 rows below its original position
# The first 2 rows will contain NaN values
df.shift(2)
```

```python
df
```

```python
# Create a new column 'Prev value' that contains the previous row's 'Price' value
# The shift(1) function moves values down by 1 position, creating a lag of 1 row
df['Prev value'] = df['Price'].shift(1)
df  # Display the dataframe with the new column
```

```python
# Calculate the price change by subtracting the previous value from the current price
# This creates a new column 'Price Change' in the dataframe
df['Price Change'] = df['Price'] - df['Prev value']
df  # Display the dataframe with the new column
```

```python
# Calculate the 5-day percentage return
# Formula: ((Current Price - Price 5 days ago) / Price 5 days ago) * 100
# This creates a new column showing how much the price has changed over a 5-day period as a percentage
df['5 Day % Return'] =( df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```
