```python
import pandas as pd

# Create a list of dates in different formats
dates = [
    '2017-01-05 2:30:00 PM',  # Date with time in 12-hour format
    'Jan 5, 2017 14:30:00',    # Date with month name and 24-hour time
    '01/05/2017',              # Date in MM/DD/YYYY format
    '2017.01.05',              # Date with dots as separators
    '2017/01/05',              # Date with slashes as separators
    '20170105'                 # Date as a continuous string without separators
]
# Convert all date formats to pandas datetime objects
# pd.to_datetime() automatically detects and parses various date formats
pd.to_datetime(dates)
```

```python
# Convert string '05/01/2017' to datetime format
# dayfirst=True means the date is in DD/MM/YYYY format (so this is January 5, 2017, not May 1)
pd.to_datetime('05/01/2017', dayfirst=True)
```

```python
# Convert string '05$01$2017' to datetime using custom format
# %d = day (05), %m = month (01), %Y = year (2017)
# The $ character is used as the separator in the date string
pd.to_datetime('05$01$2017',format = "%d$%m$%Y")
```

```python
# Convert Unix timestamp (in seconds) to a pandas datetime object
# 1769566858 represents a specific point in time (July 27, 2026)
t = 1769566858
pd.to_datetime(t, unit='s')
```
