```python
import pandas as pd
import numpy as np

# Fixed 'Dataframe' to 'DataFrame' (capital 'F')
# Added missing comma between "delhi" and "banglore"
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "humidity": [80, 60, 78],
        "temperature": [32, 45, 30]
},index = [1,2,3])
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "humidity": [68, 65, 75],
     "temperature": [21, 14, 35]
},index = [4,5,6])
us_weather  # Corrected variable name to match the definition above
```

```python
# Concatenate the india_weather and us_weather dataframes vertically into a single dataframe
df = pd.concat([india_weather,us_weather])
# Display the combined dataframe
df
```

```python
# Concatenate india_weather and us_weather dataframes into a single dataframe
# ignore_index=True resets the index of the combined dataframe to avoid duplicate indices
df = pd.concat([india_weather,us_weather],ignore_index = True)
df  # Display the concatenated dataframe
```

```python
# Concatenate india_weather and us_weather DataFrames vertically
# The 'keys' parameter adds a hierarchical index with 'India' and 'Usa' labels
# This creates a multi-index DataFrame where the first level identifies the source DataFrame
df = pd.concat([india_weather,us_weather],keys = ['India','Usa'])
df  # Display the concatenated DataFrame with hierarchical index
```

```python
# Temperature DataFrame
temperature_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30]
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["mumbai", "delhi",'banglore'],
    "windspeed": [7, 12,9]
})
windspeed_df
```

```python
# Combine temperature and windspeed dataframes horizontally (side by side)
# axis=1 means concatenation happens along columns (not rows)
pd.concat([temperature_df, windspeed_df], axis=1)
```

```python
# Create a pandas Series with weather events and name it "event"
s = pd.Series(['Humid','Dry','Rain'] , name = "event")
# Display the Series
s
```

```python
# Concatenate the temperature_df DataFrame with series s horizontally (along columns)
pd.concat([temperature_df,s],axis = 1)
```
