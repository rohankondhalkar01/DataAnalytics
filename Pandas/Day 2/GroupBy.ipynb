```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Import numpy library for numerical operations
import numpy as np 
# Read CSV file containing weather data by cities into a pandas DataFrame
df = pd.read_csv('/Users/rohankondhalkar/Desktop/Pandas/weather_by_cities.csv')
# Display the DataFrame contents
df
```

```python
# Group the DataFrame 'df' by the 'city' column
# This creates a GroupBy object that can be used for aggregation operations on city-based groups
g = df.groupby('city')
```

```python
# Loop through each group in the groupby object 'g'
# 'city' contains the group key (city name)
# 'city_df' contains the DataFrame subset for that city
for city, city_df in g:
    print(city)  # Print the name of the current city
    print(city_df)  # Print the DataFrame containing data for the current city
```

```python
# Get the subset of data where the group key equals 'mumbai'
# This returns a DataFrame containing only rows from the original DataFrame
# that belong to the 'mumbai' group
g.get_group('mumbai')
```

```python
g.max()  # Find the maximum value in the Series or DataFrame 'g'
```

```python
g.min()
```

```python
# Generate descriptive statistics for the DataFrame 'g'
# This includes count, mean, std, min, 25%, 50%, 75%, max for each numeric column
g.describe()
```

```python
# Calculate the mean of all numeric columns in the DataFrame 'g'
# The numeric_only=True parameter ensures only numeric columns are considered
g.mean(numeric_only=True)
```
