```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Import numpy library for numerical operations
import numpy as np
# Read CSV file from the specified path into a pandas DataFrame
df = pd.read_csv('/Users/rohankondhalkar/Desktop/Pandas/weatherNegative.csv')
# Display the DataFrame contents
df
```

```python
# Replace all instances of -99999 with NaN (Not a Number) values in the dataframe
# This is a common technique to handle missing or placeholder values
new_df = df.replace(-99999,np.nan)
# Display the new dataframe with NaN values instead of -99999
new_df
```

```python
# Create a new DataFrame by replacing specific values in the original DataFrame:
# - Replace -99999 with NaN (missing values)
# - Replace the string '0' with 'Sunny'
new_df = df.replace({
    -99999:np.nan,
    '0':'Sunny'
})
new_df  # Display the new DataFrame with replaced values
```

```python
# Replace any alphabetic characters in 'temperature' and 'windspeed' columns with spaces
# This helps clean the data by removing text that might be mixed with numeric values
# The regex=True parameter enables regular expression pattern matching
new_df = new_df.replace({
    'temperature': '[A-Za-z]',  # Match any uppercase or lowercase letters in temperature column
    'windspeed': '[A-Za-z]',    # Match any uppercase or lowercase letters in windspeed column
}," ",regex = True)
new_df
```

```python
# Import pandas library for data manipulation and analysis
import pandas as pd

# Create a DataFrame with two columns: 'score' and 'student'
# The DataFrame contains evaluation scores for 5 students
df = pd.DataFrame({
    'score': ['exceptional', 'average', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'julian', 'erica']
})

# Display the DataFrame
df
```

```python
# Replace categorical values with numerical values for easier analysis
# 'poor' -> 1, 'average' -> 2, 'good' -> 3, 'exceptional' -> 4
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df  # Display the transformed dataframe with numerical values
```

```python
# Convert categorical score values to numerical values
# Map text ratings to a 1-4 scale:
# - 'poor' → 1
# - 'average' → 2
# - 'good' → 3
# - 'exceptional' → 4
df['score'] = df['score'].map({
    'poor': 1,
    'average': 2,
    'good': 3,
    'exceptional': 4
})

# Display the dataframe with the updated score column
df
```
