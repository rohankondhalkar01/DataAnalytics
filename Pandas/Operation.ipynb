```python
# Import pandas library for data manipulation and analysis
import pandas as pd

# Read CSV file into a pandas DataFrame
# The file path is specified as '/Users/rohankondhalkar/Desktop/Pandas/stock_data.csv'
# The 'na_values' parameter specifies strings that should be recognized as NaN/NA values
df = pd.read_csv('/Users/rohankondhalkar/Desktop/Pandas/stock_data.csv',na_values = ['n.a.','not available']);

# Display the DataFrame
df
```

```python
# Read CSV file with custom NA values for specific columns
# - 'eps' column: 'not available' will be treated as NA
# - 'revenue' column: -1 will be treated as NA
# - 'people' column: both 'not available' and 'n.a.' will be treated as NA
# - 'price' column: both 'not available' and 'n.a.' will be treated as NA
df = pd.read_csv('/Users/rohankondhalkar/Desktop/Pandas/stock_data.csv',  na_values={
        'eps': ['not available'],
        'revenue': [-1],
        'people': ['not available','n.a.'],
        'price':['not available','n.a.']
    })
df  # Display the dataframe
```
