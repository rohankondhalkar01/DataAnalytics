```python
# Import pandas library for data manipulation and analysis
import pandas as pd

# Read an Excel file into a pandas DataFrame
# - Path specifies the location of the Excel file on the local machine
# - engine="xlrd" specifies the engine to use for reading Excel files
df = pd.read_excel(
    "/Users/rohankondhalkar/Desktop/Pandas/survey.xls",
    engine="xlrd"
)
# Display the DataFrame contents
df
```

```python
# Create a cross-tabulation table showing the count of individuals
# based on their nationality (rows) and handedness (columns)
pd.crosstab(df.Nationality,df.Handedness)
```

```python
# Create a cross-tabulation (contingency table) showing the relationship
# between Handedness and Sex variables in the dataframe
pd.crosstab(df.Handedness, df.Sex)
```

```python
# Create a cross-tabulation (contingency table) showing the count of individuals
# for each combination of Nationality and Sex categories in the dataframe
pd.crosstab(df.Nationality, df.Sex)
```

```python
# Create a cross-tabulation (contingency table) showing the frequency distribution
# of Age values for each Sex category in the dataframe
pd.crosstab(df.Sex,df.Age)
```
