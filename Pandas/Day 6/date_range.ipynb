```python
# Import pandas library for data manipulation and analysis
import pandas as pd
# Read the CSV file containing Apple stock data without dates into a DataFrame
df = pd.read_csv("/Users/rohankondhalkar/Desktop/Pandas/aapl_no_dates.csv")
# Display the first 5 rows of the DataFrame to preview the data
df.head()
```

```python
df.plot()
```

```python
# Create a DatetimeIndex with business day frequency ('B')
# Start date: June 1, 2017
# End date: June 30, 2017
# This will generate dates for all business days in June 2017 (excluding weekends)
rng = pd.date_range(start="6/01/2017",end = "6/30/2017",freq ="B")
rng
```

```python
# Set the DataFrame index to the range variable 'rng' and modify the DataFrame in-place
# instead of creating a new DataFrame object
df.set_index(rng,inplace = True)
```

```python
df
```

```python
# Create a horizontal bar plot from the DataFrame
# This visualizes all columns in the DataFrame as horizontal bars
df.plot(kind = 'barh')
```

```python
# Plot the 'Close' column of the dataframe as a line graph
# This visualizes the closing prices over time (assuming df is a time-series dataframe)
df.Close.plot()
```

```python
df["2017-06-01":"2017-06-21"].plot()
```

```python
# Extract a subset of the dataframe from June 1st to June 21st, 2017
# Then select the 'Close' column and calculate descriptive statistics
# (count, mean, std, min, 25%, 50%, 75%, max)
df["2017-06-01":"2017-06-21"].Close.describe()
```

```python
# Resamples the DataFrame to daily frequency ('D') and forward-fills missing values
# - asfreq() changes the time series frequency
# - method="pad" uses the last known value to fill gaps (also known as forward-fill)
df.asfreq("D", method="pad")
```

```python
# Resample the DataFrame to weekly frequency ('W')
# Using forward-fill ('pad') method to handle missing values
# This maintains the last known value for any missing data points
df.asfreq("W", method="pad")
```

```python
# Resample the DataFrame to hourly frequency ('h') and forward-fill (pad) missing values
# This maintains the last known value until a new value is encountered
df.asfreq("h", method="pad")
```

```python
# Create a business day date range starting from January 1, 2017
# with 72 periods (business days) and store it in variable 'rgn'
rgn = pd.date_range(start = "01/01/2017", periods = 72, freq = "B")
rgn  # Display the date range
```

```python
# Import NumPy library for numerical operations
import numpy as np 
# Generate random integers between 1 and 10 (exclusive) or 1-9 (inclusive)
# Creates a pandas Series with random values
# The Series has the same length as 'rgn' and uses 'rgn' as its index
ts = pd.Series(np.random.randint(1, 10, size=len(rgn)),index = rgn)
ts  # Display the time series
```
